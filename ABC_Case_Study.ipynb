```python
new_data=raw_data.copy()
```

```python
#converting the comma seprated values to list
new_data['cast']=new_data['cast'].str.split(',')
new_data['country']=new_data['country'].str.split(',')
new_data['listed_in']=new_data['listed_in'].str.split(',')
new_data['director']=new_data['director'].str.split(',')
new_data.head()
```

```python
#Unnesting the columns
new_data=new_data.explode('cast')
new_data=new_data.explode('country')
new_data=new_data.explode('listed_in')
new_data=new_data.explode('director')
```

```python
#removing unwanted spaces
new_data['country']=new_data['country'].str.strip()
new_data['director']=new_data['director'].str.strip()
new_data['cast']=new_data['cast'].str.strip()
new_data['listed_in']=new_data['listed_in'].str.strip()
```

```python
new_data=new_data.reset_index(drop=True)
new_data.head()
```

```python
new_data.shape
```

```python
new_data.duplicated().sum()
```

```python
new_data.drop_duplicates(inplace=True)
```

```python
new_data.duplicated().sum()
```

```python
#filling NaN or empty values
new_data['director'].fillna('unknown_director',inplace=True)
new_data['cast'].fillna('unknown_actor',inplace=True)
new_data['country'].fillna('unknown_country',inplace=True)
new_data['rating'].fillna('unknown_rating',inplace=True)
new_data['date_added'].fillna('2022-01-01',inplace=True)
new_data['duration'].fillna('unknown_duration',inplace=True)
```

```python
new_data['date_added']=pd.to_datetime(new_data['date_added'])
```

```python
#Extracting week no from every date and then creating new column
def applyweekno(g):
  if g>=1 and g<=8:
    weekno=1
  elif g>8 and g<=15:
    weekno=2
  elif g>15 and g<22:
    weekno=3
  else:
    weekno=4
  return weekno

new_data['week_of_month']=new_data['date_added'].dt.day.apply(applyweekno)
```

```python
new_data['month_added']=new_data['date_added'].dt.month_name()
```

```python
new_data['difference_in_years']=np.abs(new_data['date_added'].dt.year-new_data['release_year'])
```

```python
new_data.duplicated().sum()
```

```python
new_data.isna().sum()
```

```python
movies_data=new_data[new_data['type']=='Movie']
tv_data=new_data[new_data['type']=='TV Show']
```

**Q1. Find the counts of each categorical variable both using graphical and non-
graphical analysis**

```python
type_counts=new_data.groupby('type').aggregate({'title':'nunique'}).sort_values(by='title',ascending=False).reset_index()
type_counts.rename(columns={'title':'count'},inplace=True)

rating_counts=new_data.groupby('rating').aggregate({'title':'nunique'}).sort_values(by='title',ascending=False).reset_index()
rating_counts.rename(columns={'title':'count'},inplace=True)

listed_in_counts=new_data.groupby('listed_in').aggregate({'title':'nunique'}).sort_values(by='title',ascending=False).reset_index()
listed_in_counts.rename(columns={'title':'count'},inplace=True)
```

```python
print(type_counts,rating_counts,listed_in_counts,sep='\n\n\n')
```

```python
plt.figure(figsize=(8,20))
plt.subplot(6,1,1)
sns.barplot(data=type_counts,x='type',y='count')
plt.subplot(6,1,3)
plt.xticks(rotation=90)
sns.barplot(data=rating_counts,x='rating',y='count')
plt.subplot(6,1,5)
plt.xticks(rotation=90)
sns.barplot(data=listed_in_counts.head(10),x='listed_in',y='count')
```

**Q4. Find which is the best month to release the Tv-show or the movie.**

```python
movie_month= movies_data.groupby(by='month_added').aggregate({'title':'nunique'}).reset_index().sort_values(by='title',ascending=False).reset_index(drop=True)
movie_month.rename(columns={'title':'movie_count'},inplace=True)
movie_month
```

```python
tv_month = tv_data.groupby(by='month_added').aggregate({'title':'nunique'}).reset_index().sort_values(by='title',ascending=False).reset_index(drop=True)
tv_month.rename(columns={'title':'tv_count'},inplace=True)
tv_month
```

```python
tv_movie_count = movie_month.merge(tv_month,on='month_added',how='outer').sort_values(by=['movie_count','tv_count'],ascending=False).reset_index(drop=True)
tv_movie_count_grp = tv_movie_count.melt(id_vars=['month_added'],value_vars=['movie_count','tv_count'],var_name='type',value_name='count')
tv_movie_count_grp
```

```python
plt.xticks(rotation=90)
sns.barplot(data=tv_movie_count_grp,x='month_added',y='count',hue='type')
```

**Business Insight - 5:**

By analyzing above data we got to know that:

1. Most Movies and TV shows are released in july, april, december.
